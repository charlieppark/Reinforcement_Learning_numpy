```python
import numpy as np
```

```python
def converge(P : np.matrix, epoch = 100, epsilon = 1e-8):
    def diff(A : np.matrix, B: np.matrix):
        if A.shape != B.shape:
            raise Exception("Shape of two matrices are differend")
        else:
            return np.concatenate(np.power(A - B, 2)).sum()
    for i in range(epoch):
        _ = P*P
        if diff(_, P) < epsilon:
            break # convergence
        P = _
        print(P)
    return P
    
```

```python
# Weather Model

W = np.asmatrix(np.array([
    [0.8, 0.2],
    [0.4, 0.6]
]))

W = converge(W)

with np.printoptions(precision=3, suppress=True):
    print(W)
```
